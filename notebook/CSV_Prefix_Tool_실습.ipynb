## 📌 섹션 1: 프로젝트 소개

### 이 프로그램은 무엇인가요?

**CSV Prefix Tool**은 다음 작업을 자동화하는 데스크탑 프로그램입니다:

1. 엑셀·구글 시트에서 데이터를 **복사**한다
2. 프로그램 창의 표(그리드)에 **붙여넣기** 한다
3. 맨 앞에 붙이고 싶은 **접두 문자열**(Prefix)을 입력한다
4. **출력 생성** 버튼을 눌러 결합된 결과를 얻는다
5. 결과를 **파일로 저장**하거나 **클립보드에 복사**한다

### 📚 사용 기술 스택

| `tkinter` | GUI 창 만들기 | 파이썬 기본 내장 GUI 라이브러리 |
| `io` | 문자열을 파일처럼 다루기 | 파이썬 기본 내장 모듈 |

## 📌 섹션 2: CSV란 무엇인가? — `csv` 모듈 실습

### CSV (Comma-Separated Values) 란?

**CSV**는 데이터를 쉼표(`,`)로 구분해서 저장하는 텍스트 형식입니다.
엑셀, 구글 시트 등에서 가장 많이 사용하는 파일 형식 중 하나입니다.

```
이름,나이,직업
홍길동,30,개발자
김철수,25,디자이너
```

위 내용을 표로 보면:

| 이름 | 나이 | 직업 |
|------|------|------|
| 홍길동 | 30 | 개발자 |
| 김철수 | 25 | 디자이너 |

### 💡 `io.StringIO` 란?

파이썬의 `csv.reader`는 파일 객체를 받습니다.
하지만 우리는 파일 없이 **문자열**에서 바로 읽고 싶을 때가 있습니다.
`io.StringIO`는 **문자열을 파일처럼 다룰 수 있게** 해주는 도구입니다.

```python
io.StringIO("문자열")  →  마치 파일처럼 읽을 수 있는 객체
```

````python
# ── 필요한 모듈 가져오기 ──────────────────────────────────────────────────
# import: 다른 모듈(파일)의 코드를 가져오는 명령어
import csv   # CSV 파싱 모듈 (파이썬 기본 내장)
import io    # 입출력 도구 모듈 (파이썬 기본 내장)

# ── 예시: 엑셀에서 복사한 텍스트 (탭으로 구분) ───────────────────────────
# 엑셀에서 셀을 복사(Ctrl+C)하면 열 사이는 탭(\t), 행 사이는 줄바꿈(\n)으로 구분됩니다.
# \t = 탭 문자 (눈에 보이지 않는 공백의 한 종류)
# \n = 줄바꿈 문자
clipboard_text = "이름\t나이\t직업\n홍길동\t30\t개발자\n김철수\t25\t디자이너"

print("원본 클립보드 텍스트:")
print(repr(clipboard_text))   # repr(): 눈에 보이지 않는 특수 문자도 표시
print()
print("실제 출력:")
print(clipboard_text)
````

````python
# ── csv.reader로 파싱하기 ────────────────────────────────────────────────
# csv.reader: CSV 텍스트를 한 행씩 리스트로 읽어주는 도구
# delimiter: 구분자 (탭으로 구분된 경우 delimiter="\t")

reader = csv.reader(
    io.StringIO(clipboard_text),  # 문자열을 파일처럼 전달
    delimiter="\t"                # 구분자: 탭
)

# list()로 변환해야 실제 데이터를 얻습니다.
# (reader 자체는 '발전기'처럼 한 줄씩 생성하는 객체라서 바로 보이지 않음)
rows = list(reader)

print("파싱 결과 (2차원 리스트):")
print(rows)
print()

print("행별 출력:")
for i, row in enumerate(rows):
    # enumerate: 인덱스(번호)와 값을 함께 꺼내는 함수
    print(f"  {i}번 행: {row}")
````

## 📌 섹션 3: 구분자 자동 감지 알고리즘

### 🔍 왜 구분자를 자동으로 감지해야 할까요?

사용자가 데이터를 어디서 복사했느냐에 따라 구분자가 달라집니다:

| 출처 | 구분자 | 예시 |
|------|--------|------|
| 엑셀/구글 시트 복사 | 탭 (`\t`) | `A\tB\tC` |
| CSV 파일 | 쉼표 (`,`) | `A,B,C` |
| 일부 유럽 소프트웨어 | 세미콜론 (`;`) | `A;B;C` |

### 💡 감지 알고리즘 전략

```
탭으로 시도 → 2개 이상의 열이 감지되면 성공!
     ↓ 실패
쉼표로 시도 → 2개 이상의 열이 감지되면 성공!
     ↓ 실패
세미콜론으로 시도 → 2개 이상의 열이 감지되면 성공!
     ↓ 실패
줄 단위로 처리 (단일 열)
```

**왜 탭을 먼저 시도할까요?**
가장 흔한 복사 출처가 엑셀이기 때문입니다. 탭은 CSV 데이터 안에 잘 포함되지 않아서 오감지 위험이 낮습니다.

````python
import csv
import io

def detect_delimiter_and_parse(text):
    """
    클립보드 텍스트에서 구분자를 자동으로 감지해 2D 리스트로 반환.
    우선순위: 탭(\t) → 쉼표(,) → 세미콜론(;)
    
    Parameters
    ----------
    text : str
        클립보드에서 읽어온 원시 텍스트
    
    Returns
    -------
    list[list[str]]
        파싱된 2차원 리스트
    """
    # for 루프로 구분자 후보 리스트를 순서대로 시도합니다.
    for delimiter in ["\t", ",", ";"]:
        reader = csv.reader(io.StringIO(text), delimiter=delimiter)
        rows = list(reader)
        
        # 조건: 파싱 결과가 존재하고, 행이 2개 이상 OR 첫 행의 열이 2개 이상
        # → 이 구분자로 여러 열이 감지되었으므로 정답!
        if rows and (len(rows) > 1 or len(rows[0]) > 1):
            print(f"✅ 감지된 구분자: '{repr(delimiter)}'")
            return rows
    
    # 어떤 구분자로도 여러 열이 감지되지 않으면
    # → 줄 단위로 단일 열 반환 (리스트 컴프리헨션 사용)
    # 리스트 컴프리헨션: [표현식 for 변수 in 반복가능객체]
    # 예: [line for line in text.splitlines()] = 각 줄을 요소로 갖는 리스트
    print("⚠️ 구분자 감지 실패: 줄 단위로 처리합니다")
    return [[line] for line in text.splitlines()]


# ── 테스트 ──────────────────────────────────────────────────────────────
test_cases = [
    ("쉼표 구분",     "A,B,C\n1,2,3"),
    ("탭 구분",       "A\tB\tC\n1\t2\t3"),
    ("세미콜론 구분", "A;B;C\n1;2;3"),
    ("단일 열",       "한줄만\n있습니다"),
]

for name, text in test_cases:
    print(f"\n--- {name} ---")
    result = detect_delimiter_and_parse(text)
    print(f"결과: {result}")
````

## 📌 섹션 4: pandas DataFrame 기초

### 🗂️ DataFrame이란?

**DataFrame**은 pandas 라이브러리의 핵심 데이터 구조입니다.
엑셀의 시트(Sheet)처럼 **행(row)과 열(column)로 구성된 2차원 표**입니다.

```
파이썬 2D 리스트:          pandas DataFrame:
[                          
  ["홍길동", "30"],   →     이름   나이
  ["김철수", "25"],         0  홍길동   30
]                           1  김철수   25
```

### 💡 왜 DataFrame을 사용할까요?

- **필터링**: 특정 열만 선택하기 쉽다 (`df.iloc[:, [0, 2]]`)
- **변환**: CSV 문자열로 쉽게 내보낼 수 있다 (`df.to_csv()`)
- **정제**: 빈 값 처리, 데이터 타입 변환 등이 편리하다

이 프로젝트에서는 그리드의 데이터를 DataFrame으로 변환한 뒤 CSV로 출력합니다.

````python
import pandas as pd   # pandas를 pd라는 별명으로 가져오기 (관례적 이름)
import io

# ── 2D 리스트로 DataFrame 만들기 ─────────────────────────────────────────
data = [
    ["홍길동", "30", "개발자",  "서울"],
    ["김철수", "25", "디자이너", "부산"],
    ["이영희", "28", "기획자",  "대구"],
]

# pd.DataFrame(data, columns=...): 2D 리스트를 표 형태로 변환
# columns: 열 이름 목록 (없으면 0, 1, 2, ... 로 자동 설정)
df = pd.DataFrame(data, columns=["이름", "나이", "직업", "도시"])

print("DataFrame 전체:")
print(df)
print()
print(f"크기: {df.shape[0]}행 × {df.shape[1]}열")  # shape: (행수, 열수) 반환
print(f"열 이름: {list(df.columns)}")
````

````python
# ── 특정 열만 선택하기 (컬럼 필터링) ────────────────────────────────────
# df.iloc[행, 열]: 위치(인덱스)로 데이터 선택
#   - iloc는 integer location의 줄임말
#   - : (콜론)은 "모든 행"을 의미합니다
#   - [0, 2]는 0번째(이름)와 2번째(직업) 열만 선택

selected_indices = [0, 2]   # 선택할 열 번호 (0=이름, 2=직업)
filtered = df.iloc[:, selected_indices]  # 모든 행(:), 지정 열만

print("필터링 결과 (이름 + 직업만):")
print(filtered)
print()

# ── DataFrame → CSV 문자열로 변환 ────────────────────────────────────────
# to_csv(): DataFrame을 CSV 형식의 문자열/파일로 내보내기
#   header=False : 열 이름(헤더) 행을 포함하지 않음
#   index=False  : 행 번호(인덱스, 0,1,2...) 열을 포함하지 않음

buf = io.StringIO()         # 빈 문자열 버퍼(가상 파일) 생성
filtered.to_csv(buf, header=False, index=False)
csv_str = buf.getvalue()    # 버퍼에 쓴 내용을 문자열로 읽기

print("CSV 문자열로 변환:")
print(repr(csv_str))   # repr()로 줄바꿈 문자도 표시
print()
print("실제 출력:")
print(csv_str)
````

````python
# ── 빈 행 제거하기 ──────────────────────────────────────────────────────
# 실제 그리드에는 사용자가 입력하지 않은 빈 행이 있을 수 있습니다.
# 이 코드는 '모든 열이 비어있는 행'을 제거합니다.

data_with_empty = [
    ["홍길동", "30", "개발자"],
    ["",      "",   ""],       # 빈 행
    ["김철수", "25", "디자이너"],
    ["",      "",   ""],       # 빈 행
]

# 직접 빈 행 걸러내기 (프로젝트의 get_dataframe 방식)
# any(v.strip() for v in row): 행의 값 중 하나라도 공백이 아니면 True
filtered_data = [
    row for row in data_with_empty
    if any(v.strip() for v in row)   # 비어있지 않은 행만 포함
]

print("원본 (빈 행 포함):", data_with_empty)
print()
print("빈 행 제거 후:", filtered_data)
````

## 📌 섹션 5: 출력 생성 핵심 로직

### 🔧 핵심 규칙

접두 문자열(prefix)과 CSV 데이터를 합쳐 최종 출력을 만들 때는 4가지 경우가 있습니다:

```
prefix 있음 + CSV 있음  →  prefix + "\n" + CSV
prefix 있음 + CSV 없음  →  prefix
prefix 없음 + CSV 있음  →  CSV
prefix 없음 + CSV 없음  →  경고! (아무것도 없음)
```

### 💡 if / elif / else — 조건 분기

파이썬의 `if/elif/else`는 조건에 따라 다른 코드를 실행합니다.

```python
if 조건1:      # 조건1이 참이면
    ...
elif 조건2:    # 조건1은 거짓이고, 조건2가 참이면
    ...
else:          # 모든 조건이 거짓이면
    ...
```

````python
import io
import pandas as pd

def generate_output(prefix: str, df: pd.DataFrame, sep: str = ",") -> str:
    """
    접두 문자열과 DataFrame을 합쳐 최종 출력 문자열을 만든다.
    
    Parameters
    ----------
    prefix : str
        CSV 앞에 붙일 접두 문자열
    df : pd.DataFrame
        출력할 데이터가 담긴 DataFrame
    sep : str
        CSV 구분자 (기본값: 쉼표)
    
    Returns
    -------
    str | None
        최종 출력 문자열. 둘 다 없으면 None.
    """
    # strip(): 앞뒤 공백/줄바꿈 제거
    # 예: "  안녕\n" → "안녕"
    prefix = prefix.strip()
    
    # 경우 0: 둘 다 없음 → None 반환
    if not prefix and df.empty:
        print("⚠️ 경고: prefix도 없고 데이터도 없습니다!")
        return None
    
    # DataFrame → CSV 문자열
    csv_str = ""
    if not df.empty:     # df.empty: DataFrame이 비어있으면 True
        buf = io.StringIO()
        df.to_csv(
            buf,
            header=False,           # 헤더 행 제외
            index=False,            # 인덱스 열 제외
            sep=sep,                # 구분자
            lineterminator="\n",    # 줄바꿈 문자 통일
        )
        # rstrip("\n"): 끝의 불필요한 줄바꿈 제거
        csv_str = buf.getvalue().rstrip("\n")
    
    # 최종 조합 — 4가지 경우 중 해당하는 것 실행
    if prefix and csv_str:    # 경우 1: 둘 다 있음
        return prefix + "\n" + csv_str
    elif prefix:              # 경우 2: prefix만 있음
        return prefix
    else:                     # 경우 3: CSV만 있음
        return csv_str


# ── 실습: 함수 테스트 ────────────────────────────────────────────────────
data = [
    ["홍길동", "30", "개발자"],
    ["김철수", "25", "디자이너"],
]
filtered = pd.DataFrame([["홍길동", "개발자"], ["김철수", "디자이너"]])

prefix_text = "[START]\nname: garam\nage: 30\n[END]"

result = generate_output(prefix_text, filtered, sep=",")
print("=" * 40)
print(result)
print("=" * 40)
````

````python
# ── 4가지 경우 모두 테스트 ───────────────────────────────────────────────
import pandas as pd

sample_df = pd.DataFrame([["A", "1"], ["B", "2"]])
empty_df  = pd.DataFrame()   # 빈 DataFrame

cases = [
    ("prefix + CSV",  "접두문자열", sample_df),
    ("prefix만",       "접두문자열", empty_df),
    ("CSV만",          "",          sample_df),
    ("둘 다 없음",     "",          empty_df),
]

for name, prefix, df in cases:
    print(f"\n--- {name} ---")
    result = generate_output(prefix, df)
    print(f"결과: {repr(result)}")
````

with open("파일명", "w", encoding="utf-8") as f:

- 파일을 열었다가 닫는 것을 깜빡하면 데이터가 손실될 수 있습니다.
- `with` 문은 블록이 끝나면 (오류가 나더라도) **자동으로 파일을 닫아줍니다.**

## 📌 섹션 7: tkinter 기초 — 간단한 GUI 만들어보기

### 🖥️ tkinter란?

**tkinter**는 파이썬에 기본으로 포함된 GUI(그래픽 사용자 인터페이스) 라이브러리입니다.
별도 설치 없이 창, 버튼, 입력창 등을 만들 수 있습니다.

### 📦 주요 위젯(Widget)

| 위젯 | 클래스 | 설명 |
|------|--------|------|
| 창(윈도우) | `tk.Tk()` | 최상위 창 |
| 텍스트 표시 | `tk.Label` | 읽기 전용 텍스트 |
| 한 줄 입력 | `tk.Entry` | 사용자 입력창 |
| 버튼 | `tk.Button` | 클릭 가능한 버튼 |
| 여러 줄 입력 | `tk.Text` | 여러 줄 텍스트 영역 |
| 상자 | `tk.Frame` | 위젯을 담는 컨테이너 |

### 📐 배치 방법 (pack)

```python
widget.pack()              # 위에서 아래로 쌓기
widget.pack(side=tk.LEFT)  # 왼쪽부터 쌓기
widget.pack(fill=tk.X)     # 가로 방향으로 꽉 채우기
widget.pack(pady=10)       # 위아래 여백 10픽셀
```

### ⚠️ Jupyter에서의 주의사항

Jupyter Notebook에서 `root.mainloop()`를 실행하면 **커널이 멈춥니다.**
창을 닫아야 다음 셀을 실행할 수 있습니다.
꼭 실행할 때만 실행하고, 반드시 창을 닫은 뒤 다음 셀로 넘어가세요!

````python
# ⚠️ 이 셀을 실행하면 tkinter 창이 열립니다!
# 창을 닫아야 노트북 다음 셀을 실행할 수 있습니다.

import tkinter as tk

# ── 이벤트 핸들러 함수 ───────────────────────────────────────────────────
# 버튼을 클릭했을 때 실행될 함수입니다.
# tkinter에서 버튼 동작은 'command' 옵션으로 함수를 연결합니다.
def show_hello():
    # entry.get(): Entry 위젯의 현재 텍스트를 가져오기
    name = entry.get()
    if name.strip():   # 이름이 비어있지 않으면
        # label.config(...): Label의 속성을 변경 (text 업데이트)
        label.config(text=f"안녕하세요, {name}님! 👋")
    else:
        label.config(text="이름을 입력해주세요!")


# ── 최상위 창 만들기 ─────────────────────────────────────────────────────
# tk.Tk(): 최상위 윈도우(창)를 만드는 클래스
root = tk.Tk()
root.title("나의 첫 번째 GUI")   # 창 제목 표시줄 텍스트
root.geometry("320x200")         # 창 크기: 가로300×세로150픽셀
root.resizable(True, True)       # 창 크기 조절 가능 여부 (가로, 세로)

# ── 위젯 추가 ────────────────────────────────────────────────────────────
# tk.Label: 텍스트 표시 위젯
tk.Label(root, text="이름을 입력하세요:", font=("맑은 고딕", 11)).pack(pady=8)

# tk.Entry: 한 줄 텍스트 입력 위젯
# width: 입력창 너비 (문자 단위)
entry = tk.Entry(root, width=22, font=("맑은 고딕", 11))
entry.pack(pady=5)

# tk.Button: 클릭 가능한 버튼 위젯
# command: 버튼 클릭 시 호출할 함수 (괄호 없이 함수 자체를 넘김!)
# command=show_hello (O) / command=show_hello() (X, 즉시 실행됨)
tk.Button(
    root,
    text="인사하기",
    command=show_hello,           # 버튼 클릭 시 show_hello 함수 실행
    font=("맑은 고딕", 11),
    bg="#3BA0D8",                  # 배경색: 파랑
    fg="white",                   # 글자색: 흰색
    padx=10,
).pack(pady=8)

# 결과를 표시할 Label (처음엔 비어있음)
label = tk.Label(root, text="", font=("맑은 고딕", 12), fg="#063D6A")
label.pack(pady=5)

# ── 메인 루프 실행 ───────────────────────────────────────────────────────
# mainloop(): 이벤트 루프 시작
# 이 함수를 호출하면 창이 화면에 표시되고,
# 사용자 입력(클릭, 키보드 등)을 기다리는 무한 루프에 진입합니다.
# 창을 닫으면 mainloop()가 종료됩니다.
root.mainloop()
````

### 💡 이벤트 바인딩 (Event Binding)

버튼 클릭 외에도 다양한 이벤트에 함수를 연결할 수 있습니다.

```python
# 형식: widget.bind("<이벤트>", 함수)

entry.bind("<Return>", lambda e: show_hello())  # Enter 키
entry.bind("<FocusIn>", lambda e: print("포커스!"))  # 클릭해서 선택
root.bind("<Control-s>", lambda e: save_file())  # Ctrl+S
```

**lambda란?** 이름 없는 작은 함수입니다.
```python
lambda e: show_hello()  # e 인자를 받아 show_hello()를 호출하는 함수
```
tkinter 이벤트 핸들러는 항상 `event` 인자를 받으므로, `lambda e: ...` 형태로 씁니다.

````python
# ⚠️ 이 셀을 실행하면 tkinter 창이 열립니다! 창을 닫은 후 다음 셀 실행하세요.

import tkinter as tk

# Enter 키로도 버튼을 동작시키는 예제
def show_hello():
    name = entry.get()
    label.config(text=f"안녕하세요, {name or '이름 없음'}님!")

root = tk.Tk()
root.title("이벤트 바인딩 예제")
root.geometry("320x180")

tk.Label(root, text="이름:", font=("맑은 고딕", 11)).pack(pady=6)
entry = tk.Entry(root, width=22, font=("맑은 고딕", 11))
entry.pack(pady=4)

# Entry에 Enter 키 이벤트 바인딩
# lambda e: show_hello() → e(이벤트 객체)를 받아 show_hello()를 호출
entry.bind("<Return>", lambda e: show_hello())

tk.Button(root, text="인사하기 (또는 Enter)",
          command=show_hello, font=("맑은 고딕", 10)).pack(pady=6)

label = tk.Label(root, text="", font=("맑은 고딕", 12), fg="#063D6A")
label.pack(pady=4)

root.mainloop()
````

#### 퀴즈 3️⃣

`super().__init__(...)`는 언제 사용하나요?

A) 자식 클래스에서 부모 클래스의 생성자를 호출할 때  
B) 모든 파이썬 파일 맨 위에 써야 하는 필수 코드  
C) 파일을 저장할 때  
D) 변수를 초기화할 때  

````python
# 정답 확인: super() 사용 예시

class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal 생성: {name}")

class Dog(Animal):   # Dog는 Animal을 상속
    def __init__(self, name, breed):
        # super().__init__: 부모(Animal)의 __init__ 호출
        super().__init__(name)
        self.breed = breed
        print(f"Dog 생성: {name}, 품종: {breed}")

dog = Dog("멍멍이", "진돗개")
print(f"이름: {dog.name}, 품종: {dog.breed}")
print("→ 정답: A")
````

#### 퀴즈 5️⃣

파일을 `with open("file.txt", "w", encoding="utf-8") as f:` 로 열 때,
`"w"` 모드의 의미는?

A) 파일을 읽기 전용으로 열기  
B) 파일을 추가 모드로 열기 (기존 내용 뒤에 추가)  
C) 파일을 쓰기 모드로 열기 (없으면 생성, 있으면 덮어쓰기)  
D) 파일을 바이너리 모드로 열기  

````python
# 정답 확인: 파일 열기 모드 정리
modes = {
    "r":  "읽기 전용 (파일이 없으면 오류)",
    "w":  "쓰기 모드 (없으면 생성, 있으면 덮어쓰기)",
    "a":  "추가(append) 모드 (기존 내용 뒤에 이어쓰기)",
    "rb": "바이너리 읽기 (이미지, 동영상 등)",
    "wb": "바이너리 쓰기",
}

print("파일 열기 모드 정리:")
for mode, desc in modes.items():
    print(f"  '{mode}': {desc}")

print("\n→ 정답: C")
````
